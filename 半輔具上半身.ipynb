```python
# 0731 threading ok
from PIL import ImageFont, ImageDraw, Image
import cv2
import mediapipe as mp
import numpy as np
import random
import time
import pyttsx3
import threading

engine = pyttsx3.init()
lock = threading.Lock()
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 

def countDown(countdown_seconds, workoutNum):
    times = 0
    def say_workout():
#         try:
#             thread = threading.Thread(target=say_workout)
        lock.acquire()
        workout = ['肩膀上舉', '手肘彎曲、放平', '手指彎曲、伸直']
        engine.say("現在的動作是" + workout[workoutNum] + " 請站到能讓全身入鏡的位置 以便辨識動作")
        engine.runAndWait()
        lock.release()
#         finally:
            
        
    
    workout = ['肩膀上舉', '手肘彎曲、放平', '手指彎曲、伸直']
    start_time = time.time()
    
    while True:
        ret, frame = cap.read()
        
        # 计算倒计时数字
        elapsed_time = time.time() - start_time
        remaining_time = max(countdown_seconds - int(elapsed_time), 0)
        font = cv2.FONT_HERSHEY_SIMPLEX
        
#       倒數計時start
        # 選擇字體和大小
        from PIL import ImageFont
        fontpath = './simsun.ttc' # <== 這裡是宋体路徑
        font = ImageFont.truetype(fontpath, 48)
        # 在图像上显示倒计时数字
        text = f"即將開始: {remaining_time}"
        # 使用 PIL 在視頻幀上繪製中文字符
        img_pil = Image.fromarray(frame)
        draw = ImageDraw.Draw(img_pil)
        draw.text((width//2-120, height//2-200), text, font=font, fill=(255,255,255,0))
        frame = np.array(img_pil)
#       倒數計時end

    
#       目前的動作start
        # 使用 PIL 在視頻幀上繪製中文字符
        img_pil = Image.fromarray(frame)
        draw = ImageDraw.Draw(img_pil)
        draw.text((width//2-100, height//2), workout[workoutNum], font=font, fill=(255,255,255,0))
        frame = np.array(img_pil)
#       目前的動作end
        
        # 在窗口中显示倒计时图像
        cv2.imshow('Mediapipe Feed', frame)
        out.write(frame)
        
        
        if remaining_time >= 5 and times == 0:
            
            # 使用另一個執行緒來說話
            thread = threading.Thread(target=say_workout)
            thread.start()
            times = 1
            
                

        # 如果倒计时结束，退出循环
        if remaining_time == 0:
            break

        # 每帧显示时间间隔
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

def shoulderLift():
    global cap  # 增加 global 宣告
    # 初始化摄像头
    cap = cv2.VideoCapture(0)

    # 初始化 Mediapipe 相关
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Curl counter variables
    counter = 0
    stage = None

    ## Setup mediapipe instance
    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
        
# 在 shoulderLift 函數中直接調用 countDown 函數
        countdown_seconds = 5
        countDown(countdown_seconds, 0)
        
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            # Recolor image to RGB
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False

            # Make detection
            results = pose.process(image)

            # Recolor back to BGR
            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            # Extract landmarks
            try:
                landmarks = results.pose_landmarks.landmark

                # Get coordinates
                shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
                elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
                wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]

                # Calculate angle
                angle = calculate_angle(shoulder, elbow, wrist)

                # Visualize angle
                cv2.putText(image, str(angle),
                            tuple(np.multiply(elbow, [640, 480]).astype(int)),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

                # Curl counter logic
                if angle > 160 and angle < 175 and wrist[1] > shoulder[1]:
                    stage = "down"
                    times = 0

                if angle > 160 and angle < 175 and stage == 'down' and wrist[1] < shoulder[1]:
                    stage = "up"
                    counter += 1
                    print(counter)

            except:
                pass

            # Render curl counter
            # Setup status box
            cv2.rectangle(image, (0, 0), (265, 95), (245, 117, 16), -1)

            # Rep data
            cv2.putText(image, 'REPS', (15, 12),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
            cv2.putText(image, str(counter),
                        (10, 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)

            # Stage data
            cv2.putText(image, 'STAGE', (95, 12),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
            cv2.putText(image, stage,
                        (90, 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)
            

            # Render detections
            mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                      mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                      mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2)
                                      )

            cv2.imshow('Mediapipe Feed', image)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

# main
# 初始化影片寫入器和攝像頭
cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_filename = 'countdown_output.avi'
out = cv2.VideoWriter(output_filename, fourcc, 20.0, (width, height))

# call
shoulderLift()

# 釋放影片寫入器
out.release()
# 釋放攝像頭，關閉窗口
cap.release()
cv2.destroyAllWindows()
```
